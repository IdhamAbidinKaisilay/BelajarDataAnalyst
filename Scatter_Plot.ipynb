```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.max_columns=None
plt.style.use('ggplot')
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 16: Pandas - Visualisasi Data/MISSION+13+FILES/dummy_ojol_transactions.csv'
df = pd.read_csv(path, sep=';')
df.head(3)
```

#Scatter Plot dengan menggunakan Pandas
Scatter Plot biasa digunakan untuk mencari hubungan antara 2 variabel atau 2 parameter


```python
df.info()
```

```python
# karena type data distance adalah object maka kita perlu ubah menjadi float
# dan karena nilainya menggunakan koma maka perlu diubah menjadi titik.
df['distance'] = df['distance'].str.replace(',','.')
df['distance'] = df['distance'].astype('float')
```

```python
# buat figure (kanvas) dan axes (sumbu) â€” kita akan menggambar beberapa plot pada sumbu yang sama
fig, ax = plt.subplots()

# pilih baris yang mode-nya 'BIKE' lalu gambar scatter plot (distance vs transaction_amount_total)
# argumen ax=ax -> pakai sumbu yang sudah dibuat di atas (jadi titik BIKE dan CAR muncul di plot yang sama)
# label='BIKE' akan dipakai pada legend, color='green' memberi warna titik
df[df['mode']=='BIKE'].plot(kind='scatter', x='distance', y='transaction_amount_total', color='green', label='BIKE', ax=ax)

# pilih baris yang mode-nya 'CAR' lalu gambar scatter plot pada sumbu yang sama
# karena menggunakan ax=ax, plot ini ditumpuk di atas plot BIKE sehingga mudah dibandingkan
df[df['mode']=='CAR'].plot(kind='scatter', x='distance', y='transaction_amount_total', color='red', label='CAR', ax=ax)

fig.show()
```
