#Filtering
Filtering adalah proses memilih baris data yang memenuhi kondisi tertentu. Anda dapat melakukan filtering menggunakan berbagai operator logika.

```python
import pandas as pd
```

```python
path_pokemon = '/content/drive/MyDrive/Data Analyst 2025 + SQL/Section 9 [MISSION 6] Pandas Analisis Data/MISSION+6+FILES/Pokemon.csv'
df_pokemon = pd.read_csv(path_pokemon, index_col='Num')
df_pokemon.head()
```

##1 Filtering DataFrame

```python
filter = df_pokemon['Name'] == 'Pikachu'
df_pokemon[filter]
```

```python
filter_type_fire = df_pokemon['Type 1'] == 'Fire'
df_pokemon[filter_type_fire]
```

```python
filter_total = df_pokemon['Total'] > 525
df_pokemon[filter_total]
```

```python
filter_attack = df_pokemon['Attack'] == df_pokemon['Attack'].max()
df_pokemon[filter_attack]
```

```python
filter_attack = df_pokemon['Attack'] == df_pokemon['Attack'].min()
df_pokemon[filter_attack]
```

```python
filter_attack = df_pokemon['Attack'] < df_pokemon['Defense']
df_pokemon[filter_attack].sort_values(by='Defense', ascending=False).head()
```

```python
filter_legendary = df_pokemon['Legendary'] == True
df_pokemon[filter_legendary]
```

```python
filter1 = df_pokemon['Type 1'] == 'Ice'
filter2 = df_pokemon['Total'] > 500
df_pokemon[filter1 & filter2] # -> & (op. AND), -> | (op. OR)
```

```python
df_pokemon[filter1 | filter2] # -> & (op. AND), -> | (op. OR)
```

##2 Filtering DataFrame menggunakan Query

```python
df_pokemon.query('Name == "Pikachu"')
```

```python
df_pokemon.query("`Type 1` == 'Fire'") #menggunaka backtick (`) jika kolomnya memiliki spasi
```
