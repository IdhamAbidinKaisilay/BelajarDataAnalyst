# Analisis Data Perokok di Indonesia 2021–2024  

Analisis ini membahas **persentase perokok di Indonesia pada periode 2021–2024** dengan menggunakan **Python** dan **library Pandas**.  

Indonesia dikenal sebagai salah satu negara dengan jumlah perokok terbesar di dunia. Kebiasaan merokok tidak hanya ditemui pada usia dewasa, tetapi juga mulai dari remaja hingga lanjut usia.  

**Data**  
Data yang digunakan dalam analisis ini adalah persentase perokok di Indonesia berdasarkan kelompok umur pada tahun 2021 hingga 2024 yang bersumber dari **Badan Pusat Statistik (BPS)**.  

```python
# pd.options.display.max_rows = 10
df
```

##1. Persentase perokok tertinggi di tahun 2021–2024

```python
# Persentase perokok tertinggi di tahun 2021–2024

# Ambil data persentase tertinggi tiap tahun
persentase_tertinggi = df.loc[df.groupby('Tahun')['Persentase Perokok'].idxmax()]
persentase_tertinggi = persentase_tertinggi.reset_index()                           # reset index
persentase_tertinggi.index += 1                                                     # ubah index mulai dari 1
persentase_tertinggi[['Tahun', 'Umur', 'Persentase Perokok']]
```

##2. Persentase perokok terendah di tahun 2021–2024

```python
# Persentase perokok terendah di tahun 2021–2024

# Ambil data persentase terendah tiap tahun
persentase_terendah = df.loc[df.groupby('Tahun')['Persentase Perokok'].idxmin()]
persentase_terendah = persentase_terendah.reset_index()                            # reset index
persentase_terendah.index += 1                                                     # ubah index mulai dari 1
persentase_terendah[['Tahun', 'Umur', 'Persentase Perokok']]
```

##3. Tabel persentase perokok per tahun 2021-2024

```python
#Tabel versi 1 - persentase perokok per tahun (2021-2024)

# Pivot table: baris = umur, kolom = tahun, nilai = persentase perokok
pivot_df_v1 = df.pivot(index='Umur', columns='Tahun', values='Persentase Perokok')

# Tambahkan kolom rata-rata persentase perokok tiap umur
pivot_df_v1['Rata-rata'] = pivot_df_v1.mean(axis=1).round(2)

# Tampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pivot_df_v1
```

```python
#Tabel versi 2 - persentase perokok per tahun (2021-2024)

# Pivot table: baris = tahun, kolom = umur, nilai = persentase perokok
pivot_df_v2 = df.pivot(index='Tahun', columns='Umur', values='Persentase Perokok')

# Tambahkan baris rata-rata persentase perokok tiap umur
pivot_df_v2.loc['Rata-rata'] = pivot_df_v2.mean(axis=0).round(2)

# Tampilkan semua baris dan kolom
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pivot_df_v2
```

##4. Perubahan tahunan setiap umur (dalam persen)

```python
perubahan_tahunan_v1 = pivot_df_v1.diff(axis=1).round(2)
perubahan_tahunan_v1
```

```python
perubahan_tahunan_v2 = pivot_df_v2.diff(axis=0).round(2)
perubahan_tahunan_v2
```
