# Latihan Boxplot

Sumber data: https://www.kaggle.com/datasets/ankitbatra1210/diabetes-dataset

Key Features:
- Target: Jenis diabetes atau kondisi pradiabetes.
- Genetic Markers: Indikator kecenderungan genetik terhadap diabetes.
- Autoantibodies: Adanya autoantibodi yang umumnya dikaitkan dengan diabetes autoimun.
- Family History: Informasi apakah diketahui ada riwayat keluarga menderita diabetes.
- Environmental Factors: Detail tentang pengaruh lingkungan yang mungkin berkontribusi terhadap diabetes.
- Insulin Levels: Mengukur kadar insulin pada pasien.
- Age and BMI: Informasi demografi termasuk usia dan Indeks Massa Tubuh.
- Physical Activity and Dietary Habits: Faktor gaya hidup yang dapat mempengaruhi risiko diabetes.

## 2. Buatlah boxplot dari `Blood_Pressure` berdasarkan jenis diabetesnya/prediabeticnya (`Target`)!

```python
# Ketikan kode anda disini:
jenis_diabetes = df['Target'].unique()
data_diabetes = []

for jenis in jenis_diabetes:
    filter = df['Target'] == jenis
    data = df[filter]['Blood Pressure']
    data_diabetes.append(data)

plt.figure(figsize=(8, 6))


plt.boxplot(data_diabetes, vert=False)
plt.title('Boxplot Blood Pressure \nBerdasarkan Jenis Diabetes / Kondisi Prediabetes')
plt.ylabel('Jenis Diabetes/Kondisi Prediabetes')
plt.yticks([i for i in range(1, 14)], jenis_diabetes)
plt.grid(axis='x')

plt.show()
```
