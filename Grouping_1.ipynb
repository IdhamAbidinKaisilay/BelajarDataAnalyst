#Grouping

```python
import pandas as pd
```

```python
path_pokemon = '/content/drive/MyDrive/Data Analyst 2025 + SQL/Section 9 [MISSION 6] Pandas Analisis Data/MISSION+6+FILES/Pokemon.csv'
df = pd.read_csv(path_pokemon, index_col = 'Num')
df.head()
```

```python
filter = df['Type 1'] == 'Grass'
df_filtered = df[filter]
df_filtered.head()
```

```python
df_filtered.mean(numeric_only = True)
```

```python
grouped_df = df.groupby(by = 'Type 2')
grouped_df.max(numeric_only = True)
# grouped_df.max(numeric_only = True)['Attack']
```

```python
grouped_df = df.groupby(by = ['Type 1', 'Type 2'])
grouped_df.mean(numeric_only = True)['Speed']
```

```python
grouped_df.max(numeric_only = True)[['HP', 'Attack', 'Defense', 'Speed']]
```
