```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
plt.style.use('ggplot')
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 16: Pandas - Visualisasi Data/MISSION+13+FILES/dummy_ojol_transactions.csv'
df = pd.read_csv(path, sep=';')
df.head(4)
```

#Histogram

##Cara-cara membuat histogram dengan Pandas dan Matplotlib

###Cara 1 membuat histogram dengan Pandas dan Matplotlib

```python
df.plot.hist(column='transaction_amount_total',bins=20)

plt.title('Histogram Total Transaksi Ojol')
plt.xlabel('Total Transaksi')
plt.ylabel('Frekuensi')

plt.show()
```

###Cara 2 membuat histogram dengan Pandas dan Matplotlib


```python
df['transaction_amount_total'].plot.hist(bins=20, color='lime')

plt.title('Histogram Total Transaksi Ojol')
plt.xlabel('Total Transaksi')
plt.ylabel('Frekuensi')

plt.show()
```

###Cara 3 membuat histogram dengan Pandas dan Matplotlib

```python
plt.figure(figsize=(10,5))
df['transaction_amount_total'].plot(kind='hist', bins=20, color='lightblue')

plt.title('Histogram Total Transaksi Ojol')
plt.xlabel('Total Transaksi')
plt.ylabel('Frekuensi')

plt.show()
```

##Visualisasi Data berbasis Objek

```python
plot_total = df['transaction_amount_total'].plot(kind='hist', bins=20, color='lightblue', figsize=(10,5))

plot_total.set_title('Histogram Total Transaksi Ojol')
plot_total.set_xlabel('Total Transaksi')
plot_total.set_ylabel('Frekuensi')

plt.show()
```

#KDE Plot

```python
df.plot(kind='kde', y=['amount_delivery','amount_merchant','transaction_amount_total'])
```

```python
df['mode'].value_counts().plot(kind='barh', width=0.9, color=['lightgreen', 'y','lightblue', 'pink'])
```

#Box Plot

```python
df.plot(kind='box', y='amount_delivery', figsize = (5,5))
```

```python
# contoh data plot yang tidak akurat karena banyak nilai yang tidak masuk akal
df.plot(kind='box', y='amount_merchant', figsize = (5,5))
plt.show()
```

```python
df2 = df[df['mode'].isin(['FOOD', 'SHOP'])]
df3 = df[df['mode'].isin(['BIKE', 'CAR'])]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))

df2.plot(kind='box', y='amount_delivery', title='Ongkos Pengantaran Barang/Makanan', ax=ax1)
df2.plot(kind='box', y='amount_merchant', title='Harga Barang/Makanan', ax=ax2)
df3.plot(kind='box', y='amount_delivery', title='Ongkos Pengantaran Orang', ax=ax3)

plt.tight_layout()
plt.show()
```
