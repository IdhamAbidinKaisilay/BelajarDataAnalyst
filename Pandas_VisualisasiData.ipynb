```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
plt.style.use('ggplot')
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 16: Pandas - Visualisasi Data/MISSION+13+FILES/dummy_ojol_transactions.csv'
df = pd.read_csv(path, sep=';')
df.head(4)
```

#Histogram

##Cara 1 membuat histogram dengan pyploat

```python
df.plot.hist(column='transaction_amount_total',bins=20)

plt.title('Histogram Total Transaksi Ojol')
plt.xlabel('Total Transaksi')
plt.ylabel('Frekuensi')

plt.show()
```
