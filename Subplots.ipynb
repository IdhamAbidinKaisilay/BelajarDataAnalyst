# Subplots

Subplot dalam Matplotlib adalah cara untuk menampilkan beberapa plot dalam satu figure. Ini sangat berguna ketika Anda ingin membandingkan beberapa grafik atau menampilkan berbagai bagian dari data secara bersamaan.

KPM: https://www.bps.go.id/id/statistics-table/3/TWt0MVNGZFdiV2RaYTFoS1oyRnRSVTFOYUhSc1VUMDkjMw==/jumlah-keluarga-penerima-manfaat--kpm--dan-anggaran-bantuan-sosial-pangan-menurut-provinsi.html?year=2022

Jumlah penduduk: https://sulut.bps.go.id/id/statistics-table/2/OTU4IzI=/jumlah-penduduk-menurut-provinsi-diindonesia.html

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 15: Matplotlib: Subplots dan Subplot2grid/MISSION 12 FILES/KPM+Tahun+2022.csv'
df = pd.read_csv(path)
df.head()
```

```python
# Misi dimulai: Siapkan kanvas besar (fig) untuk dua grafik (ax) bersebelahan.
# plt.subplots(1, 2, ...) = Kita buat 1 baris, 2 kolom
# ...figsize=(10, 10) = Ukuran 10x10.
# ..., sharey=True) = Y-axis disamakan agar provinsi sejajar (sharey=True)
fig, ax = plt.subplots(1, 2, figsize=(10, 10), sharey=True)

# Grafik Kiri (ax[0]): Tampilkan Jumlah Orang.
# Gunakan barh (horizontal bar). Nilai dibagi 100.000 agar unitnya menjadi 'Juta Orang'.
ax[0].barh(df['Provinsi'], df['Realisasi Jumlah']/100000)

# Grafik Kanan (ax[1]): Tampilkan Anggaran.
# Gunakan barh (horizontal bar). Warnanya hijau. Nilai dibagi 1 Miliar agar unitnya menjadi 'Miliyar Rupiah'.
ax[1].barh(df['Provinsi'], df['Realisasi Anggaran']/1000000000, color='green')

# Beri nama yang jelas pada grafik kiri.
ax[0].set_title('Realisasi Jumlah KPM')
# Beri nama yang jelas pada grafik kanan.
ax[1].set_title('Realisasi Anggaran KPM')

# Label sumbu X kiri: "Juta Orang"
ax[0].set_xlabel('Realisasi Jumlah (Juta Orang)')
# Label sumbu X kanan: "Miliyar Rupiah"
ax[1].set_xlabel('Realisasi Anggaran (Miliyar Rupiah)')

# Judul Utama Figure: Teks besar di atas semua grafik. Atur posisi (x) dan ukuran font.
fig.suptitle('DATA KPM TAHUN 2022', fontsize=16, x=0.615)
# fig.tight_layout(): Rapatkan semua komponen grafik agar rapi dan tidak bertumpuk.
fig.tight_layout()

# fig.show(): Tampilkan hasil visualisasi kita. Selesai!
fig.show()
```

```python
#Provinsi di Pulau Jawa
provinsi_jawa = [
    "DKI Jakarta",
    "Jawa Barat",
    "Banten",
    "Jawa Tengah",
    "DI Yogyakarta",
    "Jawa Timur"
]

# Provinsi di Pulau Kalimantan
provinsi_kalimantan = [
    "Kalimantan Barat",
    "Kalimantan Tengah",
    "Kalimantan Selatan",
    "Kalimantan Timur",
    "Kalimantan Utara"
]

# Provinsi di Pulau Sumatera
provinsi_sumatera = [
    "Aceh",
    "Sumatera Utara",
    "Sumatera Barat",
    "Riau",
    "Kepulauan Riau",
    "Jambi",
    "Bengkulu",
    "Sumatera Selatan",
    "Lampung",
    "Kepulauan Bangka Belitung"
]

# Provinsi di Pulau Sulawesi
provinsi_sulawesi = [
    "Sulawesi Utara",
    "Gorontalo",
    "Sulawesi Tengah",
    "Sulawesi Selatan",
    "Sulawesi Tenggara"
]
```

```python
# Misi Subplot Dimulai:
# plt.subplots(2,2,...) = Kita buat 2 baris dan 2 kolom (Total 4 grafik).
# figsize=(12,10) = Atur ukuran kanvas figure.
# fig, ((ax1, ax2), (ax3, ax4)) = Teknik unpacking untuk memberi nama setiap subplot (Kiri-Atas, Kanan-Atas, Kiri-Bawah, Kanan-Bawah).
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize=(12,10))

# Grafik Kiri-Atas (ax1): Visualisasi Anggaran Jawa.
# ax1.barh(...) = Membuat Bar Chart Horizontal.
# df.loc[...] = Melakukan filtering data (Pandas) sebelum visualisasi.
# df['Provinsi'].isin(provinsi_jawa) = Filter/ambil baris di df yang kolom 'Provinsi'-nya ada dalam list provinsi_jawa.
# 'Realisasi Anggaran' = Data yang digunakan sebagai nilai bar (sumbu X).
ax1.barh(provinsi_jawa, df.loc[df['Provinsi'].isin(provinsi_jawa), 'Realisasi Anggaran'], color='lightblue')
ax2.barh(provinsi_kalimantan, df.loc[df['Provinsi'].isin(provinsi_kalimantan), 'Realisasi Anggaran'], color='lightgreen')
ax3.barh(provinsi_sumatera, df.loc[df['Provinsi'].isin(provinsi_sumatera), 'Realisasi Anggaran'], color='pink')
ax4.barh(provinsi_sulawesi, df.loc[df['Provinsi'].isin(provinsi_sulawesi), 'Realisasi Anggaran'], color='#FFCC66')

fig.suptitle("Penerimaan KPM 2022 \nDi Pulau-Pulau Besar Indonesia", fontsize=24)
fig.tight_layout()
fig.show()
```
