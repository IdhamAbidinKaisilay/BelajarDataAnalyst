#Data Error / Tidak Sesuai
Data tidak sesuai tergolong nilai data yang salah karena pengetikan, nilai data yang tidak wajar dan kolom-kolom yang menjadi satu.

```python
import pandas as pd
```

```python
path = '/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 10 Pandas: Pembersihan Data/MISSION+7+FILES/biodata_error.csv'
df = pd.read_csv(path)
df
```

```python
df.describe()
```

```python
filter_umur = (df['Umur'] < 0) | (df['Umur'] > 100)
df[filter_umur]
```

```python
index_umur = df[filter_umur].index
df.loc[index_umur, 'Umur'] = 40
df
```

```python
df['Jenis Kelamin'].unique()
```

```python
df['Jenis Kelamin'].value_counts()
```

```python
filter_jk = df['Jenis Kelamin'] == 'Alien'
df[filter_jk]
```

```python
index_jk = df[filter_jk].index
df.loc[index_jk, 'Jenis Kelamin'] = 'Laki-laki'
df['Jenis Kelamin'].value_counts()
```

```python
df['Jenis Kelamin'] = df['Jenis Kelamin'].str.replace('laki-laki', 'Laki-laki')
df['Jenis Kelamin'] = df['Jenis Kelamin'].str.replace('perempuan', 'Perempuan')
df['Jenis Kelamin'].value_counts()
```
