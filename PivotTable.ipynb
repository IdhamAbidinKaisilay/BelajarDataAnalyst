```python
import pandas as pd
```

```python
path_pokemon = '/content/drive/MyDrive/Data Analyst 2025 + SQL/Section 9 [MISSION 6] Pandas Analisis Data/MISSION+6+FILES/Pokemon.csv'
df = pd.read_csv(path_pokemon, index_col='Num')
df.tail()
```

```python
df.pivot_table(index='Type 1', values='Total', aggfunc='max')
```

```python
df.pivot_table(index= ['Type 1', 'Type 2'], values=['Total', 'Attack', 'Defense'], aggfunc=['max', 'min'])
```

```python
df.pivot_table(index='Type 1', columns='Stage', values='Attack', aggfunc='mean')
```
