# Latihan Seaborn: Plot Distribusi dan Plot Kategorikal

## 1. Buatlah dataframe dengan nama `mpg` dari dataset bawaan seaborn dengan nama yang sama!

```python
# Ketikan kode anda disini:
import seaborn as sns
import matplotlib.pyplot as plt

mpg=sns.load_dataset('mpg')
mpg.head(3)
```

```python
# Hasil output yang diinginkan:
```

## 2. Buatlah histogram dari kolom `mpg` dengan menyertakan garis kde!

```python
# Ketikan kode anda disini:
plt.figure(figsize=(6.4,5))
sns.histplot(mpg['mpg'],kde=True)

plt.title('Histogram MPG')
plt.xlabel('MPG')
plt.ylabel('Total Jumlah')
plt.show()
```

```python
# Hasil output yang diinginkan:
```

## 3. Buatlah boxplot dari `mpg` berdasarkan tahun modelnya!

```python
# Ketikan kode anda disini:
sns.boxplot(x='model_year', y='mpg', data=mpg)
```

```python
# Hasil output yang diinginkan:
```

## 4. Buatlah pairplot dari kolom `mpg`, `horsepower`, dan `acceleration` berdasarkan `origin`nya!

```python
# Ketikan kode anda disini:
# cara 1
# sns.pairplot(mpg, hue='origin', vars=['mpg', 'horsepower', 'acceleration'], palette='viridis')

# cara 2
sns.pairplot(data=mpg[['mpg', 'horsepower', 'acceleration', 'origin']], hue='origin', palette='viridis')
```

## 5. Buatlah barplot rata-rata  `mpg` setiap tahun berdasarkan asalnya!

```python
# Ketikan kode anda disini:
plt.figure(figsize=(10,5))
sns.barplot(x='model_year', y='mpg', hue='origin', data=mpg)

plt.title('Rata-rata MPG Dari Tahun 1970-1982\nBerdasarkan Asal Negaranya')
plt.xlabel('Tahun')
plt.ylabel('MPG')
```
