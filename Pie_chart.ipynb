#Pie chart sederhana

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('tableau-colorblind10')
path='/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 13 Matplotlib: BPS, Kaggle dan Pie Chart/shanghai_ranking_2024.csv'
df=pd.read_csv(path)

# pd.options.display.max_rows=None
df.iloc[0:10]
# df.head()
```

```python
target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

plt.pie(data, labels=label)

plt.title(f"Nilai Indikator Shanghai Ranking \ndi {target}")

plt.show()
```

#Modifikasi Pie Chart

```python
target = 'Massachusetts Institute of Technology (MIT)'
filter = df['University_Name'] == target
data = df[filter].iloc[0][3:]
label = df[filter].iloc[0][3:].index

warna = ['gold', '#DDFF51', 'lightcoral', 'lightskyblue', 'pink', '#00AAEE']
param_explode = [0, 0, 0.1, 0, 0, 0.2]
# explode = untuk menarik keluar irisan pie tertentu agar lebih menonjol
# nilai 0 = menempel, semakin besar nilainya makin jauh dari pusat

plt.pie(data, labels=label,
        colors=warna, autopct="%.1f%%",
        pctdistance=0.7, labeldistance=1.1,
        shadow=True, explode= param_explode)

plt.title(f"Nilai Indikator Shanghai Ranking \ndi {target}")

plt.show()
# autopct="%.1f%%" = menampilkan angka persentase
# "%.1f" = format angka dengan 1 digit di belakang koma
# "%%"  = tanda persen literal
# gabungan jadi "%.1f%%"  contoh 23.456  "23.5%"
# Catatan: kalau ingin satuan lain, tinggal ganti.
# misalnya "Kg"  autopct="%.1fKg"
# contoh: 23.456  "23.5Kg"
```
