#Mempersiapkan Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('tableau-colorblind10')
path='/content/drive/MyDrive/Data Analyst 2025 + SQL + Excel/Section 13 Matplotlib: BPS, Kaggle dan Pie Chart/shanghai_ranking_2024.csv'
df=pd.read_csv(path)
df
```

```python
df.info()
```

```python
filter_topnational = df['National/Regional Rank'] == '1'
df_top_universities = df[filter_topnational].head(15).sort_values('Alumni', ascending=False)
df_top_universities
```

```python
x = df_top_universities['University_Name']
y = df_top_universities['Alumni']

plt.figure(figsize=(15,5))

plt.title("Nilai Alumni Shanghai Ranking dari Universitas Terbaik Setiap Negara")
plt.xlabel("Alumni")
plt.ylabel("Universitas")

# menambahkan label pada bar
for index, value in enumerate(y):
  plt.text(value+1, index, str(value), color='black', va='center')

plt.barh(x,y, color='#77DDBB')
plt.show()
```
